# PCA

```python
import scipy.stats as sps
import numpy as np
import matplotlib.pyplot as plt

from mlalgos.unsupervised_learning.pca import PCA
```

```python
# Generate some data
dist = sps.multivariate_normal(mean = [1,1.5], cov=np.array([[2.25, 0.5],[1,1.7]]))
N = 500
X = dist.rvs(N)

# Centre the data
xbar = np.mean(X, axis=0)
X_centred = X - xbar
```

```python
# Plot the original data
fig, ax = plt.subplots(figsize=(2.8, 2.8))
ax.set_aspect('equal', 'datalim')
c = X.T[0]
ax.scatter(X.T[0], X.T[1], s=5, c=c, cmap=plt.get_cmap('jet'))
ax.axhline(0, c='k', lw=0.5)
ax.axvline(0, c='k', lw=0.5)
plt.show()
```

```python
# Perform the PCA, retaining both components
pca = PCA(2)
X_transformed = pca.fit(X_centred)
```

```python
# Plot the transformed data
fig, ax = plt.subplots(figsize=(2.8, 2.8))
ax.set_aspect('equal', 'datalim')
ax.scatter(X_transformed.T[0], X_transformed.T[1], s=5, c=c, cmap=plt.get_cmap('jet'))
ax.axhline(0, c='k', lw=0.5)
ax.axvline(0, c='k', lw=0.5)
plt.show()
```
